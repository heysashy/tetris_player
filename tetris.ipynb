```python
import pyautogui as pyauto
import cv2 
import matplotlib.pyplot as plt
import numpy as np 
# https://www.lumpty.com/amusements/Games/Tetris/tetris.html 
```

```python
def currpage(): 
    screenshot = pyauto.screenshot() 
    screenshot = cv2.cvtColor(np.array(screenshot), cv2.COLOR_BGR2HSV)
    # screenshot = screenshot[150:1000, 475:975] #Top cut off 
    border_lower = np.array([10, 77, 170], np.uint8)
    border_upper = np.array([10, 77, 170], np.uint8)

    # Create the masks 
    border_mask = cv2.inRange(screenshot, border_lower, border_upper)
    contours, hierarchy = cv2.findContours(border_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    for i, contour in enumerate(contours):
        area = cv2.contourArea(contour)
        if (area > 300000):
            print("Found box")
            x, y, w, h = cv2.boundingRect(contour)
            box = screenshot[y:y+h, x:x+w] #Top cut off 
            plt.imshow(box)
            plt.show() 
            
    return box
```
